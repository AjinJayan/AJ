```
import pandas as pd
data=pd.read_csv('data_cleaned.csv')
```

```
data.head()
```

```
data.dtypes
```

```
data.isnull().sum()
```

```
x=data.drop(['Survived'],axis=1)
y=data['Survived']
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,stratify=y)
```

```
y_train.value_counts(normalize=True)
```

```
y_test.value_counts(normalize=True)
```

```
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(random_state=42)
```

```
clf.fit(x_train,y_train)
```

```
train_score=clf.score(x_train,y_train)
train_score
```

```
test_score=clf.score(x_test,y_test)
test_score
```

OVEr fitting

```
pred_test=clf.predict(x_test)
pred_test
```

```
pred_proba=clf.predict_proba(x_test)
pred_proba
```

```
tem=clf.predict_proba(x_test)[:,1]
```

```
y_new=[]
for i in range(len(tem)):
  if tem[i] > 0.4:
    y_new.append(1)
  else:
    y_new.append(0)
```

```
y_new
```

```
from sklearn.metrics import accuracy_score
s=accuracy_score(y_new,y_test)
```

```
s
```

NOT MUCH IMPROVEMENT


```
train_acc=[]
test_acc=[]
for i in range(1,10):
  clr=DecisionTreeClassifier(random_state=42,max_depth=i)
  clr.fit(x_train,y_train)
  s_train=clr.score(x_train,y_train)
  s_test=clr.score(x_test,y_test)
  train_acc.append(s_train)
  test_acc.append(s_test)

```

```
score=pd.DataFrame({'s_train':train_acc,'s_test':test_acc,'max_depth':range(1,10)})
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.figure(figsize=(10,6))
plt.plot(score['max_depth'],score['s_train'],marker='o')
plt.plot(score['max_depth'],score['s_test'],marker='o')
plt.xlabel=['max_depth']
plt.ylabel=['performance']
plt.legend()
```

```
train_acc=[]
test_acc=[]
for i in range(2,20):
  clr=DecisionTreeClassifier(random_state=42,max_leaf_nodes=i,max_depth=5)
  clr.fit(x_train,y_train)
  s_train=clr.score(x_train,y_train)
  s_test=clr.score(x_test,y_test)
  train_acc.append(s_train)
  test_acc.append(s_test)
```

```
score=pd.DataFrame({'s_train':train_acc,'s_test':test_acc,'max_depth':range(2,20)})
```

```
plt.figure(figsize=(20,10))
plt.plot(score['max_depth'],score['s_train'],marker='o',label='train')
plt.plot(score['max_depth'],score['s_test'],marker='o')
plt.xlabel=['max_depth']
plt.ylabel=['performance']
plt.legend()
```

```
train_acc=[]
test_acc=[]
for i in range(2,20):
  clr=DecisionTreeClassifier(random_state=42,max_leaf_nodes=19,max_depth=5,min_samples_split=i)
  clr.fit(x_train,y_train)
  s_train=clr.score(x_train,y_train)
  s_test=clr.score(x_test,y_test)
  train_acc.append(s_train)
  test_acc.append(s_test)
```

```
score=pd.DataFrame({'s_train':train_acc,'s_test':test_acc,'max_depth':range(2,20)})
```

```
plt.figure(figsize=(20,10))
plt.plot(score['max_depth'],score['s_train'],marker='o',label='train')
plt.plot(score['max_depth'],score['s_test'],marker='o')
plt.xlabel=['max_depth']
plt.ylabel=['performance']
plt.legend()
```

```
train_acc=[]
test_acc=[]
for i in range(2,20):
  clr=DecisionTreeClassifier(random_state=42,max_leaf_nodes=19,max_depth=5,min_samples_split=2,min_samples_leaf=i)
  clr.fit(x_train,y_train)
  s_train=clr.score(x_train,y_train)
  s_test=clr.score(x_test,y_test)
  train_acc.append(s_train)
  test_acc.append(s_test)
```

```
score=pd.DataFrame({'s_train':train_acc,'s_test':test_acc,'max_depth':range(2,20)})
```

```
plt.figure(figsize=(20,10))
plt.plot(score['max_depth'],score['s_train'],marker='o',label='train')
plt.plot(score['max_depth'],score['s_test'],marker='o')
plt.xlabel=['max_depth']
plt.ylabel=['performance']
plt.legend()
```
